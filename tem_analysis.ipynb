```python
from crowd_evaluation import EvaluatorC
from datasets import TemDataset
import numpy as np
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
dataset = TemDataset('/Users/julianassmann/Documents/Bachelor-Thesis/Code/datasets/ent_tem/temp.standardized.tsv',
                            debug=True)
evalc = EvaluatorC(dataset, debug=True)
```

```python
ps, confs = evalc.evaluate_workers(dataset.workers, 0.05)
```

```python
ps, confs = np.minimum(ps, 1), np.minimum(confs, .5)
```

```python
min = 10
max = 20
ind = dataset.workers[min:max]
x_labels = np.char.add(np.full(ind, "W"), np.char.mod('%d', ind + 1))
fig = go.Figure()
fig.add_trace(go.Bar(y=dataset.get_measured_error_rates_for_workers(dataset.workers[min:max])*100, name=r'$p_{true}$', x=x_labels))
# fig.add_trace(go.Bar(y=ps*100, name=r'$p_{est}$', x=x_labels))
fig.add_trace(go.Bar(y=ps[min:max]*100, name=r'$p_{est}$', error_y=dict(type='data', array=confs[min:max]*100), x=x_labels))
fig.update_layout(barmode='group', title='Measured and estimated error rates in %')
fig.show()
```

```python
from evaluation_utils.utils import calculate_estimation_accuracy

calculate_estimation_accuracy(dataset.get_measured_error_rates_for_workers(dataset.workers[min:max]), ps[min:max], confs[min:max])
```

```python
dataset = TemDataset('/Users/julianassmann/Documents/Bachelor-Thesis/Code/datasets/ent_tem/temp.standardized.tsv',
                            prefilter_mode='majority_vote',
                            debug=False)
evalc = EvaluatorC(dataset, debug=False)

accuracies = []
confidence_levels = np.arange(5, 100, step=5) / 100.0
for c in confidence_levels:
    ps, confs = evalc.evaluate_workers(dataset.workers, c)
    accuracy = calculate_estimation_accuracy(dataset.get_measured_error_rates_for_workers(dataset.workers), ps, confs)
    print(c, ':', accuracy)
    accuracies.append(accuracy)
```

```python
plt.plot(confidence_levels, accuracies, label='Evaluator')
plt.plot(confidence_levels, confidence_levels, label='optimal')
plt.xlabel('confidence level')
plt.ylabel('accuracy')
plt.title('TEM accuracy of interval vs confidence (filtered)')
plt.legend()
plt.show()
```
